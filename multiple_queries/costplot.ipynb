```python
# Using t3micro ($0.0104/s) in comparison to a 1GiB Lambda function ($0.0576/s)
# Costs for ec2 are accumulated after the instance reaches the running state and is at least one minute
# so:

import matplotlib.pyplot as plt
import numpy
ec2_ps = 0.0104
lambda_ps = 0.0576

def lambda_cost(time):
    return lambda_ps * time

def ec2_cost(time):
    if time < 60:
        return 60 * ec2_ps
    else:
        return ec2_ps * time

plt.plot([0, 60, 70], [ec2_cost(0), ec2_cost(60), ec2_cost(70)], label='ec2')
plt.plot([0, 60, 70],  [lambda_cost(0), lambda_cost(60), lambda_cost(70)], label='lambda')
#plt.plot([, 2, 3, 4])
plt.ylabel('cost in USD')
plt.xlabel('time in seconds')
plt.legend()
plt.show()
```
